```python
import pandas as pd
```

```python
#Raeding my csv in panda
data_file = "Resources/2018-19.csv"
df = pd.read_csv(data_file, encoding = "ISO-8859-1")
```

```python
#reducing my dataframe 
travel_data1819 = df[['Travel Date','Travel Type',
       'Trip Length (Days)',
       'Reason for travel']]
travel_data1819
```

```python
#grouping my dataframe by travel type to figure out potential outliers
grouped_type = travel_data1819.groupby(['Travel Type']).count()
grouped_type
```

```python
#dropping the rows that have the travel type of "Hotel" since its not a maens of transportation
travel_data1819.drop(travel_data1819.loc[travel_data1819['Travel Type']=='Hotel'].index, inplace=True)
```

```python
#confirm hotelas are dropped 
grouped_type = travel_data1819.groupby(['Travel Type']).count()
grouped_type
```

```python
#resetting index of data frame 
travel_data1819.reset_index(drop=True, inplace=True)
travel_data1819
```

```python
#renaming my columns in df 
travel_data1819.rename(columns= {'Travel Date': 'Travel_Date',
                           'Travel Type': 'Travel_Type',
                           'Trip Length (Days)':'Trip_Length_Days',
                           'Reason for travel':'Reason_For_Travel'}, inplace=True)
travel_data1819
```

```python
travel_data1819.to_csv("Resources/Cleaned1819.csv", index=False, header=True) 
```
