```python
import pandas as pd
```

```python
london_201718 = pd.read_csv('Resources/2017-18-1.csv', encoding = "ISO-8859-1")
london_201718.head()
```

```python
london1_df = london_201718[['Travel Date', 'Type', 'Trip Length Days', 'Reason for travel']].copy()
# Rename the column headers
london1_df = london1_df.rename(columns={"Type": "Travel Type"})

london1_df
```

```python
grouped_type1 = london1_df.groupby(['Travel Type']).count()
grouped_type1
```

```python
london1_df.drop(london1_df.loc[london1_df['Travel Type']=='Hotel'].index, inplace=True)
london1_df
```

```python
grouped_type1 = london1_df.groupby(['Travel Type']).count()
grouped_type1
```

```python
london1_df.reset_index(drop=True, inplace=True)
london1_df
```

```python
london1_df.rename(columns= {'Travel Date': 'Travel_Date',
                           'Travel Type': 'Travel_Type',
                           'Trip Length Days':'Trip_Length_Days',
                           'Reason for travel':'Reason_for_Travel'}, inplace=True)
london1_df
```

```python
london1_df.to_csv('Resources/London1_df.csv', index=False)

```
